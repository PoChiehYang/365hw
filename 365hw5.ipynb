```python
import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt

# Generate synthetic normal data
np.random.seed(42)
n = 100
m = 10000
true_theta, true_tau = 0, 1  # True mean and precision
x = np.random.normal(true_theta, true_tau**-0.5, size=n)

# Model 1: Normal prior for theta, Gamma prior for tau
with pm.Model() as model_1:
    theta = pm.Normal("theta", mu=0, sigma=true_tau**-0.5)
    tau_SS_prior = 1
    tau_prior_n = 1
    tau = pm.Gamma("tau", alpha=tau_SS_prior/2, beta=tau_prior_n/2)
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)
    idata_1 = pm.sample(draws=m, chains=4, tune=1000)

# Model 2: Uniform prior for theta, Exponential prior for tau
with pm.Model() as model_2:
    theta = pm.Uniform("theta", lower=-10, upper=10)
    tau = pm.Exponential("tau", lam=1)
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)
    idata_2 = pm.sample(draws=m, chains=4, tune=1000)

# Model 3: Laplace prior for theta, Half-Cauchy prior for tau
with pm.Model() as model_3:
    theta = pm.Laplace("theta", mu=0, b=1)
    tau = pm.HalfCauchy("tau", beta=1)
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)
    idata_3 = pm.sample(draws=m, chains=4, tune=1000)

```

```python
# Plotting and diagnostics for Model 1
az.plot_trace(idata_1, var_names=["theta", "tau"])
plt.suptitle("Model 1: Trace Plots")
plt.show()

az.plot_posterior(idata_1, var_names=["theta", "tau"], hdi_prob=0.95)
plt.suptitle("Model 1: Posterior Distributions")
plt.show()

az.plot_autocorr(idata_1, var_names=["theta", "tau"])
plt.suptitle("Model 1: Autocorrelation Plots")
plt.show()

az.plot_energy(idata_1)

az.summary(idata_1, var_names=["theta", "tau"], round_to=4)
```

Slice sampling is an MCMC method that generates samples from a probability distribution by defining a horizontal "slice" beneath the unnormalized density curve and sampling uniformly from it. Given a current value $ x_t $, a vertical threshold $ y $ is first drawn uniformly from $ [0, f(x_t)] $, defining a horizontal slice where $ f(x) > y $. A valid interval $[a, b]$ is then identified around $ x_t $, and a new candidate $ x_{t+1} $ is drawn uniformly from this interval. If the new point does not satisfy $ f(x) > y $, the interval is reduced, and the process is repeated until acceptance.

This method can be integrated into the Metropolis in Hastings. Using example from last week, we can directly slice sample from the joint distribution(which is proportional to full conditional) instead of performing metropolis hastings for both tau and theta. This removes the need for a proposal and the acceptence/rejection step, resulting in more efficient computation (faster convergence and less work). Below, I replaced the MH steps with the slice sampling method.


```python
import scipy.stats as stats

# Data 
n = 30
x = stats.norm(loc=0, scale=1).rvs(size=n)

# Chains and Gibbs samples
C, G = 2, 10000
theta = np.zeros([C, G])  
tau = np.zeros([C, G])    

# Initialization
theta[:, 0] = np.mean(x)  
tau[:, 0] = 1 / np.var(x)  

# Prior hyperparameters
theta0 = 0      # Prior mean for theta
tau0 = 1        # Prior precision for theta
alpha = 2       # Shape parameter for Gamma prior on tau
lam = 1         # Rate parameter for Gamma prior on tau

# Define the joint posterior 
def joint_posterior(theta, tau):
    return (tau**(n/2) * np.exp(-0.5 * tau * np.sum((x - theta)**2)) *
            np.exp(-0.5 * tau0 * (theta - theta0)**2) *
            tau**(alpha - 1) * np.exp(-lam * tau))

# Slice Sampling 
def slice_sampling(f, x0, width=1.0, max_attempts=100):
    """ Performs slice sampling given an unnormalized density function f """
    y = np.random.uniform(0, f(x0))  # Sample slice height
    x_l, x_r = x0 - width, x0 + width  # Initial interval
    
    # Expand interval until endpoints fall below y
    while f(x_l) > y:
        x_l -= width
    while f(x_r) > y:
        x_r += width

    # Sample from the slice
    for _ in range(max_attempts):
        x_new = np.random.uniform(x_l, x_r)
        if f(x_new) > y:
            return x_new  # Accept
        elif x_new < x0:
            x_l = x_new  # Shrink left
        else:
            x_r = x_new  # Shrink right
    return x0  # Return previous value if max attempts reached

# Gibbs sampling with Slice Sampling
for c in range(C):  
    for g in range(1, G):  
        
        # Slice sampling for theta
        theta[c, g] = slice_sampling(
            lambda th: joint_posterior(th, tau[c, g-1]), 
            theta[c, g-1]
        )

        # Slice sampling for tau 
        tau[c, g] = slice_sampling(
            lambda ta: joint_posterior(theta[c, g], ta) if ta > 0 else 0,  
            tau[c, g-1]
        )

# Burn-in and summary statistics
burn_in = 2000
theta_samples = theta[:, burn_in:].flatten()
tau_samples = tau[:, burn_in:].flatten()

print("Theta samples:", np.mean(theta_samples), np.std(theta_samples))
print("Tau samples:", np.mean(tau_samples), np.std(tau_samples))
```

```python
theta_all_samples = theta_samples.flatten()
tau_all_samples = tau_samples.flatten()

# Compute 95% credible intervals
theta_ci = np.percentile(theta_all_samples, [2.5, 97.5])
tau_ci = np.percentile(tau_all_samples, [2.5, 97.5])


plt.figure(figsize=(12, 6))

# Theta histogram
plt.subplot(1, 2, 1)
plt.hist(theta_all_samples, bins=30, color='skyblue', edgecolor='black', density=True, alpha=0.7)
plt.axvline(theta_ci[0], color='red', linestyle='--', label=f'2.5%: {theta_ci[0]:.4f}')
plt.axvline(theta_ci[1], color='red', linestyle='--', label=f'97.5%: {theta_ci[1]:.4f}')
plt.title('Posterior Distribution of Theta')
plt.xlabel('Theta')
plt.ylabel('Density')
plt.legend()

# Tau histogram
plt.subplot(1, 2, 2)
plt.hist(tau_all_samples, bins=30, color='lightgreen', edgecolor='black', density=True, alpha=0.7)
plt.axvline(tau_ci[0], color='red', linestyle='--', label=f'2.5%: {tau_ci[0]:.4f}')
plt.axvline(tau_ci[1], color='red', linestyle='--', label=f'97.5%: {tau_ci[1]:.4f}')
plt.title('Posterior Distribution of Tau')
plt.xlabel('Tau')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()
```
